```python
import pandas as pd
```

```python
n = pd.read_csv("Name.csv")
```

```python
from scholarly import scholarly as sc
```

```python
google_dict = {}
missed_names = []
for i,x in enumerate(n.iloc[:,0]):
    temp = sc.search_author(x)
    try:
        google_dict[x] = sc.fill(next(temp))
    except:
        missed_names.append(x)
```

```python
def process_author_gs(author_json):
    '''
        Function to process authour information from author profiles from google
        scholar.
        
        Parameters
        ----------
        author_json - author information stored in json format
        
        Returns
        -------
        author_info - author information in form of dictionary
        
    '''
    temp = {}
    
#     temp["university"] = author_json["affiliation"]
    temp["total_citation"] = author_json["citedby"]
    temp["hindex"] = author_json["hindex"]
    temp["Years active"] = int(list(author_json["cites_per_year"].keys())[-1]) - int(list(author_json["cites_per_year"].keys())[0])
    temp["No_of_publication"] = len(author_json["publications"])
    
    return temp
```

```python
ret_data = {}
```

```python
for i in google_dict.keys():
    ret_data[i] = process_author_gs(google_dict[i])
```

```python
gs_data = pd.DataFrame(ret_data).T
```

```python
gs_data.to_csv("google_scholar_data.csv")
```
