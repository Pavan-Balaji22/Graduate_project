```python
Final_data = pd.read_csv("Final_data.csv",index_col=0)
```

```python
Final_data
```

```python
def summary_template(data):
    return("""{} is a reasearcher who work in the field of {}. 
He/she is a researcher who is active in this field for the past 
{:.0f} years and his/her work has been cited for a total of {:.0f} times.
With a H-index of {} he/she has published a total of {:.0f} papers. 
He/she has attened {:.0f} events and have recieved scholarships 
and grant worth {}.""".format(data["Name"],data["Department"],data["Years active"],data["total_citation"],data["hindex"],data["No_of_publication"],data["No of Event and Conferences"],data["Awards"]))
```

```python
summary = {}
```

```python
for i in range(Final_data.shape[0]):
    summary[Final_data.iloc[i,:]["Name"]] = summary_template(Final_data.iloc[i,:])
```

```python
pd.Series(summary).to_json("Summary.json")
```
